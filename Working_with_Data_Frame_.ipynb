# Data Validation Checks (With Pandas Examples)

### 1. Checking for Missing Values

```python
# Check for missing values in each column
print("Missing values per column:")
display(df.isnull().sum())

# Check for total missing values in the DataFrame
total_missing = df.isnull().sum().sum()
print(f"\nTotal missing values in the DataFrame: {total_missing}")
```

### 2. Checking Data Types

```python
# Check the data types of each column
print("Data types of each column:")
display(df.dtypes)
```

### 3. Checking for Unique Values

```python
# Check the number of unique values in each column
print("Number of unique values per column:")
display(df.nunique())

# Get the unique values in the 'class' column
print("\nUnique values in the 'class' column:")
display(df['class'].unique())
```

### 4. Checking Value Counts for Categorical Columns

```python
# Get the value counts for the 'class' column
print("Value counts for the 'class' column:")
display(df['class'].value_counts())
```

### 5. Checking for Duplicated Rows

```python
# Check for duplicated rows
print("Number of duplicated rows:")
display(df.duplicated().sum())

# Display duplicated rows
print("\nDuplicated rows:")
display(df[df.duplicated()])
```

### 6. Checking Value Ranges for Numerical Columns

```python
# Describe numerical columns to see value ranges
print("Description of numerical columns (including range):")
display(df.describe())
```
